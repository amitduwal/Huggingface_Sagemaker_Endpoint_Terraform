```python
from PyPDF2 import PdfReader
from langchain.text_splitter import CharacterTextSplitter
import fitz
import nltk
```

```python
def get_pdf_text_pymupdf(input_pdf_path):
    text = ""
    pdf_document = fitz.open(input_pdf_path)
    for page_number in range(pdf_document.page_count):
        page = pdf_document.load_page(page_number)
        text += page.get_text()
    pdf_document.close()
    return text
```

```python
def get_pdf_text(pdf_doc):
    text = ""
    # for pdf in pdf_docs:
    pdf_reader= PdfReader(pdf_doc)
    for page in pdf_reader.pages:
        text += page.extract_text()
    return text
```

```python
test_doc = "/content/drive/MyDrive/Embeddings/Human Resource Directives.pdf"
```

```python
%%time
text = get_pdf_text(test_doc)
# PyPDF2 SLow inaccurate reads " " as \n
```

```python


text_splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size = 1000,
    chunk_overlap  = 20
)
chunks = text_splitter.split_text(text)
# \n\n returns the entire document as a single piece of text
```

```python
print(chunks)
for chunk in chunks:
  print(str(len(chunk)))
```

```python
%%time
text = get_pdf_text_pymupdf(test_doc)
```

```python
%%time

text_splitter = CharacterTextSplitter(
    separator = "\n",
    chunk_size = 1500,
    chunk_overlap  = 40
)
chunks = text_splitter.split_text(text)
#best
```
